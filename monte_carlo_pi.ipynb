```python
import random
import matplotlib.pyplot as plt
```

Generates num_samples random points within the unit square.
Checks whether each point falls inside the quarter circle (x^2 + y^2 <= 1).
Counts the number of points inside the circle.T
he ratio of points inside the circle to the total number of points, multiplied by 4, gives an estimate of Ï€.

```python
def monte_carlo_pi(num_samples):
    inside_circle = 0

    x_inside = []
    y_inside = []
    x_outside = []
    y_outside = []

    for _ in range(num_samples):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1
            x_inside.append(x)
            y_inside.append(y)
        else:
            x_outside.append(x)
            y_outside.append(y)

    pi_estimate = 4 * inside_circle / num_samples

    return pi_estimate, x_inside, y_inside, x_outside, y_outside
```

```python
num_samples = 10000
```

```python
pi_estimate, x_inside, y_inside, x_outside, y_outside = monte_carlo_pi(num_samples)
```

```python
print(f"Estimated value of pi: {pi_estimate}")
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(x_inside, y_inside, color='blue', s=1, label='Inside Circle')
plt.scatter(x_outside, y_outside, color='red', s=1, label='Outside Circle')
plt.title(f"Monte Carlo Simulation to Estimate Pi\nEstimated Value: {pi_estimate}")
plt.legend()
plt.show()
```
